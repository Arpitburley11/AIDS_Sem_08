## **Aim:** To Develop structure based social media analytics model for Social Media Network to find
*   To calculate Degree centrality, Closeness centrality, Betweenness Centrality
*   To identify the most influential node.





```python
import networkx as nx
import matplotlib.pyplot as plt
```

## Create an Instagram friends graph

```python
G = nx.Graph()

# Define nodes (Instagram users)
users = ["Alice", "Bob", "Charlie", "David", "Eve", "Frank", "Grace", "Hannah", "Ian", "Jack",
         "Kevin", "Lily", "Mason", "Nora", "Oscar", "Paul", "Quinn", "Rachel", "Steve", "Tina"]
G.add_nodes_from(users)

# Define edges (friendship connections)
friendships = [("Alice", "Bob"), ("Alice", "Charlie"), ("Bob", "David"), ("Charlie", "Eve"),
               ("David", "Eve"), ("Eve", "Frank"), ("Frank", "Grace"), ("Grace", "Hannah"),
               ("Hannah", "Ian"), ("Ian", "Jack"), ("Jack", "Alice"), ("Charlie", "Hannah"),
               ("David", "Jack"), ("Bob", "Grace"), ("Kevin", "Lily"), ("Lily", "Mason"),
               ("Mason", "Nora"), ("Nora", "Oscar"), ("Oscar", "Paul"), ("Paul", "Quinn"),
               ("Quinn", "Rachel"), ("Rachel", "Steve"), ("Steve", "Tina"), ("Tina", "Kevin"),
               ("Alice", "Nora"), ("David", "Oscar"), ("Frank", "Paul"), ("Hannah", "Quinn"),
               ("Jack", "Steve"), ("Charlie", "Tina"), ("Bob", "Mason"), ("Grace", "Rachel")]
G.add_edges_from(friendships)
```

## Visualize the graph

```python
plt.figure(figsize=(7, 5))
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_color='lightblue', edge_color='gray', node_size=1000, font_size=10)
plt.title("Instagram Friends Network")
plt.show()
```

## Compute centrality measures

```python
degree_centrality = {k: round(v, 3) for k, v in nx.degree_centrality(G).items()}
closeness_centrality = {k: round(v, 3) for k, v in nx.closeness_centrality(G).items()}
betweenness_centrality = {k: round(v, 3) for k, v in nx.betweenness_centrality(G).items()}
```

```python
# Print results in column format
print("User\t\tDegree Centrality\tCloseness Centrality\tBetweenness Centrality")
for user in degree_centrality:
    print(f"{user}\t\t{degree_centrality[user]}\t\t\t{closeness_centrality[user]}\t\t\t{betweenness_centrality[user]}")
```

```python
# Plot centrality measures
plt.figure(figsize=(14, 5))
plt.subplot(1, 3, 1)
plt.bar(degree_centrality.keys(), degree_centrality.values(), color='blue')
plt.xticks(rotation=90)
plt.title("Degree Centrality")

plt.subplot(1, 3, 2)
plt.bar(closeness_centrality.keys(), closeness_centrality.values(), color='green')
plt.xticks(rotation=90)
plt.title("Closeness Centrality")

plt.subplot(1, 3, 3)
plt.bar(betweenness_centrality.keys(), betweenness_centrality.values(), color='red')
plt.xticks(rotation=90)
plt.title("Betweenness Centrality")

plt.tight_layout()
plt.show()
```

## Find the most influential nodes

```python
most_influential_degree = max(degree_centrality, key=degree_centrality.get)
most_influential_closeness = max(closeness_centrality, key=closeness_centrality.get)
most_influential_betweenness = max(betweenness_centrality, key=betweenness_centrality.get)
```

```python
print("Most Influential User by Degree Centrality:", most_influential_degree)
print("Most Influential User by Closeness Centrality:", most_influential_closeness)
print("Most Influential User by Betweenness Centrality:", most_influential_betweenness)
```
